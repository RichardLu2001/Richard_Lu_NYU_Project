```python
import pandas as pd
import pandas_market_calendars as mcal
```

```python
nyse = mcal.get_calendar('NYSE')
trade_days = nyse.valid_days(start_date='2009-01-01', end_date='2025-07-02')
master = pd.DataFrame(index=trade_days.date)
master.index = pd.to_datetime(master.index)

master.head()
```

```python
master.index
```

```python
pickle_files = ['Baltic Dry Index.pkl','Bond ETF.pkl', 'Bonds OAS.pkl', 'CPIAUCSL.pkl','CRB Excess.pkl',
                'daily_treasury_par_yield.pkl','EFFR-target.pkl', 'EPU.pkl', 'GSCPI.pkl', 'financial stress.pkl',
                'Google Trends.pkl', 'Industry Production Index.pkl','investor survey.pkl', 'MOVE.pkl', 'PMI.pkl', 
                'Recession Sahm.pkl', 'PC_Equity.pkl', 'PC_Index.pkl','Recession.pkl','VIX.pkl', 'VVIX.pkl',
                'SPY ETF.pkl','USD_Index.pkl', 'TPU.pkl','gold.pkl','silver.pkl','oil.pkl','natural gas.pkl','soybean.pkl'
                ,'copper.pkl','Russell 2000 ETF.pkl','SP400.pkl', 'SP500.pkl', 'USD.pkl']

for pfile in pickle_files:
    
    df = pd.read_pickle(pfile)
    
    df.index = pd.to_datetime(df.index)
    df = df.sort_index()

    df_reindexed = df.reindex(master.index)

    master = master.join(df_reindexed, how='left')

master.head()
```

```python
master.drop(['1 Mo','6 Mo','1 Yr','3 Yr','7 Yr','20 Yr','EFFR'],axis=1,inplace=True)
master.info()
```

```python
master['month'] = master.index.month
```

```python
master = master.ffill()
master.to_pickle('My Data.pkl')
```

```python
df = pd.read_pickle('My Data.pkl')
df['10Y-2Y_fut10'] = df['10Y-2Y'].shift(-10)
df['2Y10Y30Y_fut10'] = df['2Y10Y30Y'].shift(-10)
df['10Y-2Y_fut21'] = df['10Y-2Y'].shift(-21)
df['2Y10Y30Y_fut21'] = df['2Y10Y30Y'].shift(-21)

df['steepen_2week'] = (df['10Y-2Y_fut10'] > df['10Y-2Y']).astype(int)
df['curvature_2week'] = (df['2Y10Y30Y_fut10'] > df['2Y10Y30Y']).astype(int)
df['steepen'] = (df['10Y-2Y_fut21'] > df['10Y-2Y']).astype(int)
df['curvature'] = (df['2Y10Y30Y_fut21'] > df['2Y10Y30Y']).astype(int)

df = df.dropna(subset=['10Y-2Y_fut10','2Y10Y30Y_fut10','10Y-2Y_fut21','2Y10Y30Y_fut21'])
df.info()
```

```python
df.drop(['10Y-2Y_fut10','2Y10Y30Y_fut10','10Y-2Y_fut21','2Y10Y30Y_fut21'],axis=1,inplace=True)
df.to_pickle('My Data.pkl')
```
